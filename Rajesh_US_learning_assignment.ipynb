#Kmeans clustering
#Using elbow method to find optimal number of clusters

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=42)
    kmeans.fit(df1)
    wcss.append(kmeans.inertia_)
plt.subplots(figsize=(20,6))
plt.plot(range(1,15),wcss)
plt.xlabel('No of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method to find the optimum number of Cluster')
plt.show()
```

As a result of constructing this graph, we see that the optimal number of clusters is either 2 or 3

K-means Clustering Model

```python
kmeans = KMeans(n_clusters=3, init='k-means++',random_state=42)
y_kmeans = kmeans.fit_predict(df1)
y_kmeans
```

```python
# Using silhouette_score for evaluation 
from sklearn.metrics import silhouette_score
```

```python
sil_avg=silhouette_score(df1,y_kmeans)
print('The Silhouette Score for Cluster number 3 is', sil_avg)
```

```python
kmeans1 = KMeans(n_clusters=2, init='k-means++',random_state=42)
y_kmeans1 = kmeans1.fit_predict(df1)
y_kmeans1
```

```python
sil_avg1=silhouette_score(df1,y_kmeans1)
print('The Silhouette Score for Cluster number 2 is',sil_avg1)
```

```python
#Better silhouette_score is obtained for the cluster number 2 
```

```python
#Using Db scan
```

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.neighbors import NearestNeighbors
```

```python
NN=NearestNeighbors(n_neighbors=2)
nn=NN.fit(df1)
distance,indices=nn.kneighbors(df1)
```

```python
db=DBSCAN(eps=40,min_samples=3)
db_model=db.fit(df1)
cluster=db_model.labels_
```

```python
#Using PCA for reducing the dimensionality of data
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler=StandardScaler()
scaled_data=scaler.fit_transform(df1)
scaled_data=pd.DataFrame(scaled_data,columns=df1.columns)
scaled_data.head()
```

```python
pca1=PCA(n_components=1)
x_pca1=pca1.fit_transform(scaled_data)
```

```python
x_pca1.shape
```

```python
pca1.components_
```
